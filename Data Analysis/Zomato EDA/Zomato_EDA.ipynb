### Import the library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
%matplotlib inline
```

### Reading the dataset

```python
data = pd.read_csv("./Zomatodataset/zomato.csv" , encoding='latin-1')
data.head()
```

### Data Cleaning

```python
# cuisines has 9 empty slots
data.isnull().sum()

# syntax to get list of columns having null values
[features for features in data.columns if data[features].isnull().sum() > 0]
```

```python
data[data.isnull()['Cuisines']]
```

```python
sns.heatmap(data.isnull() , yticklabels=False,cbar=False , cmap='viridis' )
plt.show()
```

### Merge the country column in data (DataFrame)

```python
data_country = pd.read_excel('./Zomatodataset/Country-Code.xlsx')
data_country
```

```python
data = pd.merge(data ,data_country , on='Country Code' , how='left')

data.head()
```

### Question 1 : Which country has highest no. of restaurant?

Answer : India (After that USA, UK)

```python
#  data['Country'].unique()

country = data['Country'].value_counts().index

no_of_res = data['Country'].value_counts().values

plt.pie(no_of_res[0:3] , labels=country[0:3] , autopct='%1.2f%%' , radius=1.1 , explode=[0 ,0.1,0.1])
plt.show()
```

 ### Question 2 : Number of restaurant on the basis of ratings?

Answer : 
1. Rating between 4.5 to 4.9 ---> Excellent
2. Rating between 4.0 to 4.4 ---> Very Good
3. Rating between 3.5 to 3.9 ---> Good
4. Rating between 2.5 to 3.4 ---> Average
5. Rating between 1.8 to 2.4 ---> Poor



```python
ratings = data.groupby(['Aggregate rating' , 'Rating color' , 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings.head()
```

```python
sns.barplot(x='Aggregate rating' , y='Rating Count' , data=ratings)
plt.title('No. of Restaurant (On basis of Rating)')
plt.show()
```

```python
sns.barplot(x='Aggregate rating' , y='Rating Count' , hue='Rating color', palette=['blue' ,'red' , 'orange' , 'yellow' , 'green' , 'darkgreen'], data=ratings)
plt.title('No. of Restaurant (On basis of Rating)')
plt.show()
```

### Question 3 : Show of  frequency of ratings?

```python
sns.countplot(x='Rating color' ,data=ratings , palette=['blue' ,'red' , 'orange' , 'yellow' , 'green' , 'darkgreen'])
plt.show()
```

### Question 4 : Country has most 0 rating?

Answer : India >> Brazil >> US >> UK

### Question 5: Which currency is used by which country?

```python
data.groupby(['Country','Currency' ]).size().reset_index()[['Country' , 'Currency']]
```
